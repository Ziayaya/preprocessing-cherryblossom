# Preprocessing Menggunakan Dataset Temperature and Cherry Blossom Status

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("cherry_bloom.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = "mean")
imputer.fit(X[:, 1:2])
X[:, 1:2] = imputer.transform(X[:, 1:2])
```

```python
print(X)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [("encoder", OneHotEncoder(sparse=False), [0])], remainder = "passthrough")
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

```python
from sklearn.preprocessing import LabelEncoder #Khusus untuk mengencoder label
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```
